Total Sales by Year

```python
year_total = evsales.iloc[:,2:].sum()
year_total
```

```python
#calculate year over year growth
yoy_growth = year_total.pct_change()*100
yoy_growth
```

```python
#determing the trend by overall trend by comparing first and last year
initial_value = year_total.iloc[0].sum()
final_value =  year_total.iloc[-1].sum()
overall_growth = ((final_value)/(initial_value))*100
print(f"\n overall growth b/w year {2015} and {2024} is {overall_growth}%")
```

Total Sales per Manufacturer

```python
#create a total sale columns
evsales["total_sales"] = evsales.iloc[:,2:].sum(axis = 1)
evsales
```

```python
Manufacturers_sales = evsales.groupby('Maker')['total_sales'].sum().reset_index()
Manufacturers_sales
```

```python
#SORTING Manufacturers BY TOTAL SALES in decending order
Manufacturers_sales=Manufacturers_sales.sort_values(by = 'total_sales', ascending = False)
Manufacturers_sales
```

Top 10 Manufacturers

```python
#bar representation for top ten companies
df = Manufacturers_sales.iloc[:10,:]
df.plot(kind='bar', x='Maker', y='total_sales')
plt.xlabel('Category')
plt.ylabel('total sales')
```

which company have highest market shares

```python
#calculate total sales accros all Manufacturerss
total_market_sales = Manufacturers_sales['total_sales'].sum()
total_market_sales
```

## CALCULATE MARKET SHARE FOR EACH MANUFACTURERS

```python
Manufacturers_sales['market share(%)'] = (Manufacturers_sales['total_sales'] /total_market_sales)*100
```
